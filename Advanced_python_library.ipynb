```python
import json
import time
from functools import wraps
from typing import Any, Generator
```

```python
cache = {}

def cache_result(func):
    @wraps(func)
    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    return wrapper
```

```python
def retry(times=3):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for i in range(times):
                try:
                    return func(*args, **kwargs)
                except Exception:
                    print("Retry", i+1)
            raise Exception("Maximum retries exceeded")
        return wrapper
    return decorator
```

```python
def validate(func):
    @wraps(func)
    def wrapper(value):
        if value < 0:
            raise ValueError("Negative values are not allowed")
        return func(value)
    return wrapper
```

```python
def square_numbers(numbers):
    for num in numbers:
        yield num * num

def even_numbers(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num
```

```python
class FileManager:

    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_value, traceback):
        self.file.close()
```

```python
class Registry(type):

    registry = {}

    def __new__(cls, name, bases, attrs):
        new_class = super().__new__(cls, name, bases, attrs)
        cls.registry[name] = new_class
        return new_class
```

```python
class Plugin(metaclass=Registry):
    pass

class MyPlugin(Plugin):
    pass

print("Registered Plugins:")
print(Registry.registry)
```

```python
def serialize(data):
    return json.dumps(data)

def deserialize(data):
    return json.loads(data)
```

```python
def multiply(a: int, b: int) -> int:
    return a * b
```

```python
@cache_result
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print("Factorial:", factorial(5))
```

```python
@validate
def square(x):
    return x * x

print(square(10))
```

```python
print("Squares")
for value in square_numbers([1,2,3,4,5]):
    print(value)
print()
print("Even Numbers")
for value in even_numbers([1,2,3,4,5,6,7,8]):
    print(value)
```

```python
with FileManager("sample.txt","w") as file:
    file.write("Welcome to Advanced Python Library")

with FileManager("sample.txt","r") as file:
    print(file.read())
```

```python
student = {
    "Name":"Sanskruti",
    "Age":20,
    "City":"Pune"
}

text = serialize(student)

print("Serialized")
print(text)
print()
print("Deserialized")
print(deserialize(text))
```

```python
count = 0
@retry(3)
def test():
    global count
    count += 1
    if count < 3:
        raise Exception("Error")
    return "Success"
print(test())
```
